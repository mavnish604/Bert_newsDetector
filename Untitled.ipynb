```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("csmalarkodi/isot-fake-news-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
df_true = pd.read_csv("data/True.csv")
```

```python
df_fake = pd.read_csv("data/Fake.csv")
```

```python
df_true
```

```python
df_fake
```

```python
df_true['label'] = [1 for _ in df_true['text']]
```

```python
df_true
```

```python
df_true = df_true[["text","label"]]
```

```python
import re

# Function to remove the "CITY (Reuters) - " pattern
def clean_reuters_leakage(text):
    # This regex looks for an optional city, "(Reuters)", and a hyphen at the start of the text
    return re.sub(r'^.*?\(Reuters\)\s*-\s*', '', text)

# Apply it only to the true news dataset (since fake news doesn't have it)
df_true['text'] = df_true['text'].apply(clean_reuters_leakage)

# Let's check if it worked
print(df_true['text'].head())
```

```python
df_true
```

```python
df_fake['label'] = [0 for _ in df_fake['text']]
```

```python
df_fake = df_fake[["text","label"]]
```

```python
df_fake
```

```python
df = pd.concat([df_true, df_fake], axis=0)
```

```python
df
```

```python
from sklearn.utils import shuffle
df=shuffle(df)
```

```python
X=df["text"]
y=df["label"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=42)
```

```python
from datasets import Dataset
from sentence_transformers import SentenceTransformer, SentenceTransformerTrainer, SentenceTransformerTrainingArguments
from sentence_transformers.losses import BatchHardTripletLoss
```

```python
train_dataset = Dataset.from_dict({
    "sentence": X_train.tolist(),
    "label": y_train.tolist()
})
```

```python
eval_dataset = Dataset.from_dict({
    "sentence": X_test.tolist(),
    "label": y_test.tolist()
})
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
train_loss = BatchHardTripletLoss(model=model)
```

```python
args = SentenceTransformerTrainingArguments(
    output_dir="./fake_news_model",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    eval_strategy="epoch",
    save_strategy="epoch",
)
```

```python
trainer = SentenceTransformerTrainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    loss=train_loss,
)
```

```python
trainer.train()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.decomposition import PCA
```

```python
log_history = trainer.state.log_history
df_logs = pd.DataFrame(log_history)

plt.figure(figsize=(10, 6))
if 'loss' in df_logs.columns:
    train_logs = df_logs.dropna(subset=['loss'])
    plt.plot(train_logs['step'], train_logs['loss'], label='Training Loss', color='blue')

if 'eval_loss' in df_logs.columns:
    eval_logs = df_logs.dropna(subset=['eval_loss'])
    plt.plot(eval_logs['step'], eval_logs['eval_loss'], label='Validation Loss', marker='o', color='orange')

plt.title('Training and Validation Loss Curve')
plt.xlabel('Training Steps')
plt.ylabel('Loss (Batch Hard Triplet)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Generating embeddings for Test set... (this takes a moment)")
# We use the trained model to turn text into vectors
X_train_emb = model.encode(X_train.tolist(), show_progress_bar=True)
X_test_emb = model.encode(X_test.tolist(), show_progress_bar=True)

# Fit a simple Logistic Regression classifier on the frozen embeddings
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train_emb, y_train.tolist())

# Make predictions on the test set
y_pred = clf.predict(X_test_emb)
```

```python
cm = confusion_matrix(y_test.tolist(), y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fake (0)", "True (1)"])

fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(cmap=plt.cm.Blues, values_format='d', ax=ax)
plt.title('Confusion Matrix (Logistic Regression on Embeddings)')
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_test.tolist(), y_pred, target_names=["Fake (0)", "True (1)"]))
```

```python
pca = PCA(n_components=2)
X_test_pca = pca.fit_transform(X_test_emb)

plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=X_test_pca[:, 0], 
    y=X_test_pca[:, 1], 
    hue=y_test.tolist(), 
    palette=["red", "green"], 
    alpha=0.6, 
    edgecolor=None
)
plt.title('2D PCA Visualization of Sentence Embeddings')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Label', labels=['Fake (0)', 'True (1)'])
plt.show()
```
